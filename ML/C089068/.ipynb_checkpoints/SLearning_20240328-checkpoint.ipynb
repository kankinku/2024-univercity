```python
# forge dataset

import mglearn
import matplotlib.pyplot as plt
import matplotlib

# generate dataset
X, y = mglearn.datasets.make_forge()
#plot dataset
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.legend(["Class 0","Class 1"], loc=4)
plt.xlabel("First feature")
plt.ylabel("Second feature")
print("X.shape:{}".format(X.shape))

# 26개의 데이터 피처가 2개
```

```python
# p46
import numpy as np
from sklearn.datasets import load_breast_cancer
# cancer이라는 함수 정의
cancer = load_breast_cancer()

# 클래스별 샘플이름
print("cancer.keys(): \n{}".format(cancer.keys()))

# malignant : 양성 / benign : 음성 / real data
print("클래스별 샘플 갯수: \n",format({n: v for n, v in zip(cancer.target_names, np.bincount(cancer.target))}))

```

```python
# p46
from sklearn.datasets import load_iris
# cancer이라는 함수 정의
iris_dataset = load_iris()

print("샘플 갯수: \n",format({n: v for n, v in zip(iris_dataset.target_names, np.bincount(iris_dataset.target))}))
```
